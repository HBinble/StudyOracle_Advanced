### 함수로 Oracle 연결

```python
# 오라클 연결/접속/커서 얻기
def setDsn() : 
    return cx_Oracle.makedsn("localhost", 1521, "orcl")

def setConn(dsn) : 
    return cx_Oracle.connect("busan_06", "dbdb", dsn)

def setCursor(conn) : 
    return conn.cursor()

# 오라클 커서 반납/접속 끊기
def closeOracle(cursor, conn) :
    cursor.close()
    conn.close()

# 데이터 가져오기
def gerFetchAll(sql,cursor):
    cursor.execute(sql)
    return cursor.fetchall()

# 컬럼 가져오기
def getColumn(cursor):
    col =[]
    for tp in cursor.description:
        print(tp[0])
        col.append(tp[0])
    return col

# 데이터프레임 만들기
def getDataFrame(row,col):
    return pd.DataFrame(row,columns=col)

# 최종
def getAll(sql):
    dsn = setDsn()
    conn = setConn(dsn)
    cursor = setCursor(conn)
    row = gerFetchAll(sql,cursor)
    col = getColumn(cursor)
    closeOracle(cursor, conn)
    return getDataFrame(row,col)
```
